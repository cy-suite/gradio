```
# Downloading files from the demo repo
import os
os.mkdir('mp4')
!wget -q -O mp4/en.mp4 https://github.com/gradio-app/gradio/raw/main/demo/video_with_caption/mp4/en.mp4
os.mkdir('tmp')
!wget -q -O tmp/en.srt https://github.com/gradio-app/gradio/raw/main/demo/video_with_caption/tmp/en.srt
!wget -q -O tmp/en.vtt https://github.com/gradio-app/gradio/raw/main/demo/video_with_caption/tmp/en.vtt
```

```
import gradio as gr

css = (
    "footer {display: none !important;} .gradio-container {min-height: 0px !important;}"
)

with gr.Blocks(css=css) as demo:
    # gr.Video()
    gr.VideoWithCaption(
        label="Video File Test",
        show_label=True,
        interactive=True,
        value=["mp4/en.mp4", "tmp/en.vtt"],
    )

demo.launch()
```
